```python
# changing the datatype of 'carat' from 'float' to 'int'
df['carat'] = df['carat'].astype(int)
df.info()
```

```python
df   # values of carat are now integers
```

```python
len(df)  # returns no. of rows in df
```

```python
# to add a new column 'Zeroes'
df['Zeroes'] = [i for i in range(len(df))]   # on rhs, we've provided list of vlaues of new column
df
```

```python
# creating a new column using any particular column(by applying function on it using 'apply')
def fx(a):
    return a *a

df['ones'] = df['Zeroes'] .apply(fx)
```

```python
# to save the file in required format
df.to_csv("export.csv")
# this will return the index values(when opened in excel) with the data in output file
```
